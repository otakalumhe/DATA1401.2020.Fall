```
data_path="./COVID-19/"
```

```
us_files = glob.glob(data_path + "csse_covid_19_data/csse_covid_19_daily_reports_us/*.csv")
all_files = glob.glob(data_path + "csse_covid_19_data/csse_covid_19_daily_reports/*.csv")
print(len(us_files), len(all_files))
print(all_files)
```

```
a_date_string = []
for i in range(len(all_files)):
    a_date_string.append(all_files[i].split("/")[-1].split(".")[0])
print(a_date_string)
```

```
timestamp = []
for i in range(len(all_files)):
    timestamp.append(time.mktime(datetime.datetime.strptime(a_date_string[i],
                                                            "%m-%d-%Y").timetuple()))
print(timestamp)
```

```
def SortedTime(input_list):
    diff = len(input_list) // 2
    while diff > 0:

        for i in range(diff, len(input_list)):
            temp = input_list[i]
            j = i
            # Sort the sub list for this diff

            while j >= diff and input_list[j - diff] > temp:
                input_list[j] = input_list[j - diff]
                j = j - diff
            input_list[j] = temp


        diff = diff // 2

SortedTime(timestamp)
print(timestamp)

ts_n = []
for j in range(len(timestamp)):
    timestamp_n = timestamp[j]
    dt_object =  dt.fromtimestamp(timestamp_n).strftime("%m-%d-%Y")
    ts_n.append(dt_object+'.csv')
print(ts_n)
```

```
df = []
for w in range(len(ts_n)):
    df.append(pd.read_csv(data_path+"csse_covid_19_data/csse_covid_19_daily_reports/"+ts_n[w]))
print(df)
```

```
province_country = [df[-1]['Province_State'] + df[-1]['Country_Region']]
print(province_country)
```
